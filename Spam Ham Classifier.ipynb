```python
import os
from collections import Counter
import math
```

```python
ham_doc_names = []
for names in os.listdir("enron1/ham"):
    ham_doc_names.append(names)
spam_doc_names = []
for names in os.listdir("enron1/spam"):
    spam_doc_names.append(names)
len(spam_doc_names)
```

```python
no_of_spam = len(spam_doc_names)
no_of_ham = len(ham_doc_names)
total = no_of_spam + no_of_ham
prior = []
prior.append(no_of_spam/total)
prior.append(no_of_ham/total)
prior
```

```python
ham_word_count = {}
spam_word_count = {}

for filename in os.listdir(".\enron1\\ham"):
    file = open('.\enron1\ham\\' + filename, errors='ignore')
    wordcount = Counter(file.read().split())
    for word in wordcount:
        if word not in ham_word_count:
            ham_word_count[word] = []
            ham_word_count[word].append(wordcount[word])
        else:
            ham_word_count[word][0] += wordcount[word]
        if word not in spam_word_count:
            spam_word_count[word] = [0]

for filename in os.listdir(".\enron1\\spam"):
    file = open('.\enron1\spam\\' + filename, errors='ignore')
    wordcount = Counter(file.read().split())
    for word in wordcount:
        if word not in spam_word_count:
            spam_word_count[word] = []
            spam_word_count[word].append(wordcount[word])
        else:
            spam_word_count[word][0] += wordcount[word]
        if word not in ham_word_count:
            ham_word_count[word] = [0]
```

```python
sum_spam = 0
sum_ham = 0
for word in spam_word_count:
    sum_spam += spam_word_count[word][0]
for word in ham_word_count:
    sum_ham += ham_word_count[word][0]
sum_spam
```

```python
# train classifier
for word in spam_word_count:
    x = (spam_word_count[word][0]+1)/(sum_spam + len(spam_word_count))
    spam_word_count[word].append(x)

for word in ham_word_count:
    x = (ham_word_count[word][0]+1)/(sum_ham + len(ham_word_count))
    ham_word_count[word].append(x)
```

```python
def predict(filename):
    file = open(filename)
    wordcount = Counter(file.read().split())
    score=[]
    score.append(math.log(prior[0]))
    score.append(math.log(prior[1]))
    for word in wordcount:
        if word in spam_word_count:
            score[0] += math.log(spam_word_count[word][1])
        if word in ham_word_count:
            score[1] += math.log(ham_word_count[word][1])
    if(score[0]>score[1]):
        return 0
    else:
        return 1
```

```python
ans = 0
for i in range(len(ham_doc_names)):
    if predict("enron1/ham/"+ham_doc_names[i])==1:
        ans+=1
print(ans)
```
